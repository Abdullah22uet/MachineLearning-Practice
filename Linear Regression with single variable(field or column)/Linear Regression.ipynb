<h1 style="background-color:#02317d;color:white;padding:12px 0 10px 10px;">Import Necessary Libraries</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
```

```python
df.head(4)
```

```python
df.info()
```

```python
df.rename(columns={"per capita income (US$)":"income"},inplace=True)
```

<h1 style="background-color:#02317d;color:white;padding:12px 0 10px 10px;">Visualize the dataset using Scatter Plot in Matlotlib and Seaborn. Both are same.</h1>

```python
plt.scatter(df.year,df.income,color='red',marker='+')
```

```python
import seaborn as sns
sns.scatterplot(x=df.year,y=df.income)
plt.show()
```

<h1 style="background-color:#02317d;color:white;padding:12px 0 10px 10px;">Training the model</h1>

```python
year = df.drop(["income"],axis=1)
year.head(3)
```

```python
income = df["income"]
income.head(3)
```

```python
model = LinearRegression()
model.fit(year,income)
```

<h1 style="background-color:#02317d;color:white;padding:12px 0 10px 10px;">Predictions</h1>

```python
model.predict([[2024],[2025],[2026]])
```
